## 당뇨병 여부 판단
- 예측컬럼: Outcome (0 정상, 1 당뇨) 당뇨병일 확률 예측
- 평가지표: roc-auc
- 제출파일명: result.csv (1개컬럼, 컬럼명 pred)

## 라이브러리 및 데이터 불러오기

```python
import pandas as pd

# 데이터 불러오기
train = pd.read_csv("/kaggle/input/bigdatacertificationkr/diabetes_train.csv")
test = pd.read_csv("/kaggle/input/bigdatacertificationkr/diabetes_test.csv")
train.shape, test.shape

# train.head()
train.info()
train['Outcome'].value_counts()
# print(dir(train))
# train.describe()

# print(dir(sklearn.preprocessing))
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])

train = train.drop('id', axis=1)
test = test.drop('id', axis=1)

target = train.pop('Outcome')

from sklearn.ensemble import RandomForestClassifier
# print(dir(sklearn.ensemble))

model = RandomForestClassifier()
model.fit(train, target)

# print(dir(model))
# help(model.predict_proba)

predictions = model.predict_proba(test) # 2차원 배열 반환

# print(predictions)
pred = predictions[:, 1] # 클래스 1일 확률만 뽑음 당뇨병일 확률만
# print(pred)

output = pd.DataFrame({'pred': pred})
output.head()

output.to_csv('result.csv', index=False)

result = pd.read_csv('result.csv')
print(result)
```

## 데이터 전처리

```python
# 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])
```

```python
# id 제외
train = train.drop('id',axis=1)
test = test.drop('id',axis=1)
```

```python
# 타겟
target = train.pop("Outcome")
```

## 학습 및 예측

```python
# 학습 및 예측
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=2022)
model.fit(train, target)
predictions = model.predict_proba(test)
```

```python
# 당뇨 확률값 선택
pred = predictions[:,1]
```

## 파일 생성

```python
output = pd.DataFrame({'pred': pred})
output.head()
```

```python
# 수험번호.csv로 출력
output.to_csv('result.csv', index=False)
```
